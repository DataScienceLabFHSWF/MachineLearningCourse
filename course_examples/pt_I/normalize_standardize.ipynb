```python
import numpy as np
import matplotlib.pyplot as plt
```

## Normalization
Normalization is the process of scaling individual samples to have unit norm, it is useful when the data has different scales and is uniformly distributed

```python
def min_max_normalize(data):
    min_val = min(data)
    max_val = max(data)
    normalized_data = [(x - min_val) / (max_val - min_val) for x in data]
    return normalized_data

data = [10, 20, 30, 40, 50]
data = np.random.uniform(0, 100, 100)
normalized_data = min_max_normalize(data)
print(f"Original data: {data}")
print(f"Normalized data: {normalized_data}")

print(f"Normalized data sorted: {np.sort(normalized_data)}")

plt.hist(data, bins=10, alpha=0.5, label='Original data')
plt.hist(normalized_data, bins=10, alpha=0.5, label='Normalized data')
plt.legend(loc='upper right')
plt.title('Original and normalized data')
plt.show()

plt.hist(normalized_data, bins=10, alpha=0.5, label='Normalized data')
plt.legend(loc='upper right')
plt.title('Normalized data')
plt.show()
```

## Standardization
Standardization of data is useful when the data is normally distributed and has a standard deviation that is not equal to 1.
it is used in machine learning algorithms to ensure that the data scales are similar and that the algorithm converges faster.

```python
def standardize_data(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    standardized_data = [(x - mean) / std_dev for x in data]
    return standardized_data

data = np.random.normal(0, 100, 100)
standardized_data = standardize_data(data)
print(f"Original data: {data}")
print(f"Standardized data: {standardized_data}")

print(f"Standardized data mean: {np.mean(standardized_data)}")
print(f"Standardized data std deviation: {np.std(standardized_data)}")

plt.hist(data, bins=20, alpha=0.5, label='Original data')
plt.hist(standardized_data, bins=20, alpha=0.5, label='Standardized data')
plt.legend(loc='upper right')
plt.title('Original and standardized data')
plt.show()

plt.hist(standardized_data, bins=20, alpha=0.5, label='Standardized data')
plt.legend(loc='upper right')
plt.title('Standardized data')
plt.show()
```
